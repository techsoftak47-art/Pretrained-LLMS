```python
import os
import pdfplumber
from groq import Groq
from google.colab import drive

# 1. Mount Google Drive
try:
    drive.mount('/content/drive', force_remount=False)
except:
    pass

# 2. Initialize Groq Client with your key
GROQ_API_KEY = "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"
client = Groq(api_key=GROQ_API_KEY)

def extract_text_from_pdf(pdf_path):
    """Extracts text from abc.pdf on Drive."""
    if not os.path.exists(pdf_path):
        return f"Error: File not found at {pdf_path}. Ensure 'abc.pdf' is in your main Drive folder."

    try:
        with pdfplumber.open(pdf_path) as pdf:
            text = ""
            for page in pdf.pages:
                content = page.extract_text()
                if content:
                    text += content + "\n"
        return text
    except Exception as e:
        return f"Error reading PDF: {str(e)}"

def generate_mcqs(text, num_questions=5):
    """Generates MCQs using the correct Llama 3.3 model and response format."""
    truncated_text = text[:12000]

    prompt = f"""
    Create {num_questions} high-quality Multiple Choice Questions (MCQs) based on the text below.
    Format:
    Q: [Question]
    A) [Option]
    B) [Option]
    C) [Option]
    D) [Option]
    Answer: [Letter]

    Text:
    {truncated_text}
    """

    try:
        completion = client.chat.completions.create(
            messages=[
                {"role": "system", "content": "You are a professional academic quiz generator."},
                {"role": "user", "content": prompt}
            ],
            model="llama-3.3-70b-versatile",
            temperature=0.5,
        )
        # Correctly accessing the first choice's message content
        return completion.choices[0].message.content
    except Exception as e:
        return f"API Error: {str(e)}"

# --- EXECUTION ---
pdf_path_in_drive = "/content/drive/MyDrive/abc.pdf"

extracted_text = extract_text_from_pdf(pdf_path_in_drive)

if "Error" in extracted_text:
    print(extracted_text)
else:
    print("Extracting text from abc.pdf and generating MCQs...")
    result = generate_mcqs(extracted_text)

    print("\n" + "="*40)
    print("           GENERATED MCQS")
    print("="*40)
    print(result)
    print("="*40)

    # Save to Drive for convenience
    with open("/content/drive/MyDrive/mcq_results.txt", "w") as f:
        f.write(result)
    print("\n✅ Success! Questions printed above and saved to 'mcq_results.txt' in your Drive.")
```
