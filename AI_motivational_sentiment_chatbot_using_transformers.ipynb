```python
# ==========================================================
# GOOGLE COLAB AI MOTIVATIONAL SENTIMENT CHATBOT
# USING HUGGING FACE + HF_TOKEN
# ==========================================================

# Install libraries
!pip install -q transformers torch huggingface_hub

# ==========================================================
# IMPORTS
# ==========================================================

from transformers import pipeline
from huggingface_hub import login
import random

# ==========================================================
# LOGIN TO HUGGING FACE
# ==========================================================

HF_TOKEN = "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"

login(HF_TOKEN)

# ==========================================================
# LOAD SENTIMENT MODEL
# ==========================================================

model_name = "distilbert-base-uncased-finetuned-sst-2-english"

sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model=model_name
)

# ==========================================================
# MOTIVATIONAL RESPONSES
# ==========================================================

positive_responses = [
    "😊 That's wonderful to hear! Keep going, you're doing great!",
    "🌟 Amazing! Your positive mindset will take you far.",
    "💪 Keep up the good energy and confidence!",
    "🚀 You're on the right track. Believe in yourself!",
    "👏 Great attitude! Success comes from consistency."
]

negative_responses = [
    "💙 I'm sorry you're feeling this way. Remember, difficult moments are temporary.",
    "🌈 Every challenge helps you grow stronger. You can overcome this.",
    "💪 Don't give up. You are more capable than you think.",
    "✨ Tough times never last forever. Better days are ahead.",
    "🤝 Be kind to yourself. Progress takes time and patience.",
    "🔥 Failure is not the end — it's part of learning and growing.",
    "🌻 Keep moving forward one step at a time. You're doing better than you realize."
]

neutral_responses = [
    "🙂 Thanks for sharing that with me.",
    "🌟 Stay focused and keep believing in yourself.",
    "💡 Every day is a new opportunity to improve.",
    "🚀 Keep learning and growing!"
]

# ==========================================================
# CHATBOT FUNCTION
# ==========================================================

def chatbot_response(user_text):

    # Analyze sentiment
    result = sentiment_pipeline(user_text)

    label = result[0]['label']
    score = result[0]['score']

    # Generate motivational reply
    if label == "POSITIVE":
        reply = random.choice(positive_responses)

    elif label == "NEGATIVE":
        reply = random.choice(negative_responses)

    else:
        reply = random.choice(neutral_responses)

    return label, score, reply

# ==========================================================
# START CHATBOT
# ==========================================================

print("======================================")
print(" AI Motivational Sentiment Chatbot ")
print("======================================")
print("Type 'exit' to stop.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: 🌟 Remember — you are capable of amazing things. Goodbye!")
        break

    label, score, reply = chatbot_response(user_input)

    print(f"\nBot: {reply}")
    print(f"Sentiment: {label}")
    print(f"Confidence: {score:.4f}\n")
```
