```python
# ============================================
# AI Email Reply Generator (Google Colab)
# Updated Groq Models - 2026
# ============================================

# Install Groq SDK
!pip install -q groq

from groq import Groq
from getpass import getpass

# ============================================
# ENTER GROQ API KEY
# ============================================

api_key = getpass("Enter your GROQ API KEY: ")

# Initialize Groq client
client = Groq(api_key=api_key)

# ============================================
# FUNCTION TO GENERATE EMAIL REPLIES
# ============================================

def generate_email_reply(email_content, tone="professional"):

    prompt = f"""
You are an AI Email Assistant.

Analyze the received email and generate:

1. SHORT professional reply
2. DETAILED professional reply

Tone: {tone}

Received Email:
{email_content}

Return format:

SHORT REPLY:
<reply>

DETAILED REPLY:
<reply>
"""

    response = client.chat.completions.create(

        # Updated working Groq model
        model="llama-3.3-70b-versatile",

        messages=[
            {
                "role": "system",
                "content": "You generate professional email replies."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],

        temperature=0.7,
        max_tokens=500
    )

    return response.choices[0].message.content

# ============================================
# MAIN PROGRAM
# ============================================

print("\n======================================")
print(" AI-Powered Email Reply Generator ")
print("======================================\n")

# Email Input
email = input("Paste the received email:\n\n")

# Tone Selection
print("\nChoose Tone:")
print("1 - Professional")
print("2 - Friendly")
print("3 - Formal")
print("4 - Casual")

choice = input("\nEnter choice (1-4): ")

tone_map = {
    "1": "professional",
    "2": "friendly",
    "3": "formal",
    "4": "casual"
}

tone = tone_map.get(choice, "professional")

# ============================================
# GENERATE REPLIES
# ============================================

print("\nGenerating AI email replies...\n")

try:

    result = generate_email_reply(email, tone)

    print("======================================")
    print(" GENERATED EMAIL RESPONSES ")
    print("======================================\n")

    print(result)

except Exception as e:

    print("\nERROR OCCURRED:")
    print(e)
```
