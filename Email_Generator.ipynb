```python
import gradio as gr
from groq import Groq
```

```python
GROQ_API_KEY = "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"
```

```python
client = Groq(api_key=GROQ_API_KEY)
```

```python
def generate_email(rough_points, email_type, tone):
    try:
        if rough_points is None or rough_points.strip() == "":
            return "Please enter rough points."

        prompt = f"""
You are a professional email writing assistant.

Convert the rough points below into a proper {email_type} email.

Tone: {tone}

Rough Points:
{rough_points}

Output format:
Subject:
Email Body:

Rules:
- Use clear and professional language.
- Keep the email ready to send.
- Do not add unnecessary information.
"""

        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            temperature=0.3,
            max_tokens=500
        )

        return response.choices[0].message.content

    except Exception as e:
        return "Error occurred: " + str(e)
```

```python
demo = gr.Interface(
    fn=generate_email,
    inputs=[
        gr.Textbox(
            lines=10,
            label="Rough Points",
            placeholder="Example: Apply for assistant professor role, attach resume, interested in AI/Data Science..."
        ),
        gr.Radio(
            choices=[
                "Job Application",
                "Meeting Request",
                "Leave Request",
                "Follow-up",
                "Project Update",
                "Formal Request",
                "Thank You"
            ],
            value="Job Application",
            label="Email Type"
        ),
        gr.Radio(
            choices=[
                "Formal",
                "Professional",
                "Polite",
                "Friendly",
                "Concise"
            ],
            value="Formal",
            label="Tone"
        )
    ],
    outputs=gr.Textbox(
        lines=15,
        label="Generated Email"
    ),
    title="AI Email Draft Generator",
    description="Enter rough points and generate a professional email using Groq."
)
```

```python
demo.launch(share=True, debug=True)
```
